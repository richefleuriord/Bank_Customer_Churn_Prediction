# *6-Déploiement*

## Déploiement du Modèle XGBoost

### Introduction

Le **déploiement** représente la phase finale du projet de *Data Science*, où le modèle prédictif devient un **outil opérationnel** utilisable par les décideurs.  
Dans ce projet, le modèle **XGBoost**, identifié comme le plus performant lors de la modélisation, a été intégré dans une **application web interactive** développée avec **Streamlit**.

Cette application permet aux **managers et analystes de la banque** de :
- saisir les caractéristiques d’un client ;  
- visualiser instantanément la **probabilité de départ (churn)** ;  
- et d’appuyer leurs décisions de fidélisation sur une **analyse prédictive fiable et explicable**.

L’objectif est de **transformer un modèle performant en un outil pratique, simple et accessible**, capable de générer des prédictions en temps réel.

---

### Objectifs du Déploiement

- **Accessibilité** : rendre le modèle utilisable via une interface web intuitive.  
- **Interaction** : permettre la saisie dynamique de nouveaux profils clients.  
- **Décisionnel** : afficher la probabilité de churn sous forme de pourcentage clair et visuel.  
- **Fiabilité et traçabilité** : charger automatiquement le pipeline complet et les colonnes finales utilisées à l’entraînement.  
- **Maintenance** : assurer la possibilité de mise à jour et de réentraînement du modèle à mesure que de nouvelles données clients arrivent.

---

### Méthodologie de Déploiement

#### Sauvegarde du modèle final
Le modèle **XGBoost** optimisé est sauvegardé sous le nom:

**Best_Model_Deployment.pkl**


Il s’agit d’un **pipeline complet (prétraitement + modèle)** exporté avec `joblib`.

---

#### Création de l’interface utilisateur (Streamlit)
L’application comprend :
- un **formulaire interactif** de saisie client (âge, solde, score de satisfaction, type de carte, etc.) ;  
- un **bouton de prédiction** déclenchant le calcul du churn ;  
- un **résumé visuel** de la probabilité de départ et de fidélité du client.

---

#### Préparation automatique des données utilisateur
Les entrées saisies sont transformées pour correspondre aux **21 variables finales** du modèle, grâce au fichier :

**columns_final.pkl**


Cela garantit une **cohérence parfaite** entre les données d’entraînement et les prédictions en production.

---

#### Affichage et Interprétation des Résultats
- Le résultat de la prédiction est accompagné de la **probabilité (%) de départ et de rétention**.  
- Une **alerte visuelle** (`st.warning` / `st.success`) s’affiche selon le risque détecté.  
- Des **indicateurs numériques** (`st.metric`) permettent une lecture rapide des résultats.

---

#### Maintien et supervision du modèle
L’application est conçue pour être **réutilisée et réentraînée facilement** à partir de nouvelles données.

# 🧍 Profil du Client #1



Ce client est un **homme appartenant à la catégorie des jeunes**, résidant en **France**, et entretient une **relation bancaire depuis 3 ans**, ce qui traduit une **ancienneté modérée** dans l’institution.  

Son **score de crédit de 600** reflète une **capacité de remboursement moyenne**, tandis que son **solde de compte de 5 000 €** et son **salaire estimé à 40 000 €** indiquent une situation financière **stable et correcte**.  

Il détient **un seul produit bancaire**, laissant penser qu’il n’a pas encore pleinement exploré l’ensemble des offres de la banque.  
Il possède **une carte de crédit de type Silver**, dispose d’un **crédit actif**, et est **actif dans ses opérations bancaires**, traduisant une **participation régulière à la vie financière de l’institution**.  

Avec un **score de satisfaction de 5 sur 10**, le client se situe dans une **zone moyenne de contentement**, exprimant une expérience ni pleinement positive ni négative.  
Il a cumulé **300 points de fidélité**, témoignant d’un **engagement modéré** dans les activités promotionnelles ou services de la banque.

---

## Résultat du Modèle XGBoost

Le modèle prédit que ce client est **susceptible de rester** au sein de la banque avec une **probabilité de 65.62 %**, tandis que la **probabilité de départ (churn)** est de **34.38 %**.  

**Conclusion :** Le client présente un **profil stable**, mais il existe un **risque modéré de départ** qui pourrait être réduit par des actions de fidélisation.

---

## Analyse Économique et Comportementale

Ce profil révèle un **client jeune et financièrement stable**, avec un comportement bancaire sain et une **participation active** à la vie de la banque.  
Son **activité régulière**, combinée à un **revenu correct** et à l’**utilisation d’une carte de crédit Silver**, renforce la confiance de l’institution envers lui.  

Cependant, son **score de satisfaction moyen** et la **faible diversification de ses produits** indiquent que son **attachement à la banque n’est pas encore profond**.  

Une **politique de fidélisation proactive** (offres personnalisées, conseils financiers, produits d’investissement) pourrait **accroître sa satisfaction et renforcer sa fidélité à long terme**.

---

*Interprétation stratégique :*  
Pour sécuriser la fidélité de ce client, la banque doit **augmenter sa satisfaction globale** et **proposer des produits adaptés à ses besoins**.  
L’objectif est de transformer ce client stable mais à risque modéré en un client **hautement engagé et fidèle**.
